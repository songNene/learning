# 쿼리문 작성과 데이터 필터링

## 쿼리문 작성하기(SELECT, FROM)
전체 컬럼 조회하기
- SELECT 구문에 * 기호를 사용하면 해당 테이블의 모든 컬럼을 조회할 수 있다.

```python
# bike 테이블에서 전체 컬럼 조회하기
mysql("""

SELECT * 
FROM bike

""")
```

```python
# bike테이블에서 날짜, 계절, 날씨 컬럼을 추출

mysql("""

SELECT dteday, season, weathersit
FROM bike

""")
```

## 전체 데이터 필터링하기(WHERE)

- SQL에서 쿼리 결과 필터링은 중요한 기술입니다.
- WHERE는 FROM절 다음에 위치합니다.
- WHERE절에는 연산자를 같이 사용해야 합니다.

```python
# 2012년을 기준으로 데이터를 확인하기

mysql("""

SELECT *
FROM bike
WHERE yr = 2012

""")
```

# 비교 연산자 사용하기
- 비교 연산자는 SQL 구문에서 필수적이다.
- 아래의 표 이미지 참고    
    
 ![image.png](./images/2.png)

```python
# 비교 연산자 =
# 2012년을 기준으로 날짜, 계절, 날씨 컬럼을 확인해 봅시다. LIMIT은 전체 데이터 중 100개만 확인

mysql("""  

SELECT dteday, season, weathersit
FROM bike
WHERE yr = 2012
LIMIT 100    

""")
```

```python
# =는 '같다'는 의미로 사용될 수도 있다.
# 계절이 봄인 전체 데이터 확인하기

mysql("""

SELECT *
FROM bike
WHERE season = 'spring'

""")
```

```python
# 비교 연산자 != 또는 <>
# 특정 데이터를 제외하여 선택할 때 사용
# 계절이 봄이 아닌 전체 데이터를 확인

mysql("""  

SELECT *
FROM bike
WHERE season != 'spring'
""")
```

```python
# <> 연산자 사용

mysql(""" 

SELECT *
FROM bike
WHERE season <> 'spring'
""")
```

```python
# 비교 연산자 >, >=
# 컬럼을 기준으로 데이터 값보다 큰 경우 사용할 수 있다.
# 숫자 타입의 금액이나 날짜 타입의 날짜를 데이터 값으로 사용할 때 사용

# 23시 이후 대여건에 대한 전체 데이터 확인하기
mysql("""

SELECT *
FROM bike
WHERE hr >= 23
LIMIT 100

""")
```

```python
# 비교 연산자 <, <=
# 컬럼을 기준으로 데이터 값보다 작은 경우에 사용
# 온도 0.2미만 전체 데이터 추출하기

mysql("""

SELECT *
FROM bike
WHERE temp < 0.2
LIMIT 100

""")
```

```python
# 전체 대여 건수가 100건보다 작은 데이터를 추출

mysql(""" 

SELECT *
FROM bike
WHERE cnt < 100

""")
```

# 논리 연산자 사용하기
- 여러가지 조건을 주고 싶을 때 사용
- 논리 연산자를 계속 추가하면 복잡한 조건도 쉽게 필터링할 수 있다.

![image.png](./images/3.png)


## 1. AND 연산자
- 앞의 조건과 뒤의 조건 둘 다 만족한 참인 값인 경우 필터링을 해준다.
- 조건을 계속 추가해야할 때 사용한다.

```python
# bike 테이블에서 2011년도 겨울 시즌의 데이터 조회하기
# yr = 2011 AND season = 'winter'

mysql("""

SELECT *
FROM bike
WHERE yr = 2011
    AND season = 'winter'
LIMIT 100

""")
```

```python
# bike 테이블에서 2011년도 봄 시즌의 데이터 조회하기

mysql("""

SELECT *
FROM bike
WHERE yr = 2011
    AND season = 'spring'

""")
```

```python
# bike 테이블에서 2011년도 봄시즌의 22시에 발생한 데이터

mysql(""" 

SELECT *
FROM bike
WHERE yr = 2011
    AND season = 'spring'
    AND hr = 22

""")
```

```python
# bike 테이블에서 2011년도 봄시즌의 일요일에 발생한 데이터
mysql("""

SELECT *
FROM bike
WHERE yr = 2011
    AND season = 'spring'
    AND weekday = 'Sunday'

""")
```

## 2. OR
- 앞의 조건을 만족하거나 뒤의 조건을 만족 즉, 한쪽이라도 참이면 만족시킨 값을 필터링한다.
- 조건을 계속 추가해야할 때 사용한다.

```python
# hr이 10이거나 11인 행 추출

mysql("""

SELECT *
FROM bike
WHERE hr = 10 
      OR hr = 11

""")
```

```python
# bike 테이블에서 2011년 또는 봄시즌의 데이터 출력하기
mysql("""

SELECT *
FROM bike
WHERE yr = 2011
      OR season = 'spring'

""")
```

## 3. IN
- 컬럼이 어떤 값들의 집합에 속한 경우 IN을 사용한다.
- 조회하고자 하는 데이터 값이 여러 개일 떄 사용
- =과 유사하지만 =은 데이터 값 하나를 조건으로 줄 수 있다.
- IN은 데이터 값을 여러 개의 list로 지정할 수 있다.
    - 여러 개 값의 리스트 중 하나의 값이 만족하면 만족하는 결과값을 출력한다.

```python
# N을 사용하여 1월(Jan), 3월(Mar), 5월(May)에 대한 전체 데이터를 추출

mysql("""

SELECT *
FROM bike
WHERE mnth IN ('Jan', 'Mar', 'May')

""")
```

```python
# OR로도 동일한 데이터를 추출할 수 있지만 하나의 컬럼에서 같은 인자를 조건으로 줄 때 IN을 사용하는 것이 효율적

mysql("""

SELECT *
FROM bike
WHERE mnth = 'Jan'
OR mnth = 'Mar' 
OR mnth = 'May'

""")
```

## 4. NOT
- 논리연산자의 의미 반전
- 조건을 부정으로 만들어 부정 연산자라고 한다.
- 선언한 조건에 대해 반대 결과를 반환


```python
# 월(Jan), 3월(Mar), 5월(May)이 아닌 전체 데이터를 추출

mysql("""

SELECT *
FROM bike
WHERE mnth NOT IN ('Jan', 'Mar', 'May')

""")
```

```python
# 1월(Jan), 3월(Mar), 5월(May)이 아니고, 전체 대여 건수가 10보다 큰 전체 데이터를 추출
mysql(""" 

SELECT *
FROM bike
WHERE mnth NOT IN ('Jan', 'Mar', 'May')
      AND cnt > 10

""")
```

## 5. NULL
- 값이 지정되지 않아 알 수 없는 값
- NULL은 0이나 공백이랑 다르다.

```python
# 날씨 데이터의 값이 NULL값인 데이터를 추출
mysql("""

SELECT *
FROM bike
WHERE weathersit IS NULL

""")
```

```python
# 날씨 데이터의 값이 NULL값이 아닌 전체 데이터를 추출

mysql("""

SELECT *
FROM bike
WHERE weathersit IS NOT NULL

""")
```

## 6. Between A and B
- A와 B 사이에 값이 존재할 때 사용, 주의할 점은 A와 B를 포함한다.
- 두 값의 범위에 해당하는 행을 출력할 때 사용
- 주로 날짜사이에 간격을 줄 때 자주 사용하는 편
- 최소값과 최대값을 지정한 값의 범위 내에 있는 값들을 검색하기 위해 사용
- number Between 3 AND 5와 number>=3 AND number<=5는 동일한 결과를 출력

```python
# 2011년 3월 30일 ~ 31일 사이의 데이터를 확인

mysql(""" 

SELECT *
FROM bike
WHERE dteday Between '2011-03-30' and '2011-03-31'

""")
```

```python
# 비교 연산자를 사용하여 검색

mysql("""

SELECT *
FROM bike
WHERE dteday >= '2011-03-30' 
      AND dteday <= '2011-03-31'

""")
```

# 산술 연산자 사용하기

## 산술연산자
- 데이터값을 계산할 때 사용
- 산술연산자는 +, -, *, / 등이 있다. (우선 순위: () > *, / > +,-)
- SELECT, WHERE 문에서 사용
- 문자열 혹은 날짜타입 데이터에 산술연산자를 적용하면 원하지 않는 값을 얻게될 수 있다.
- 실무에서 많이 쓰이며 비즈니스를 위한 다양한 리포트를 작성할 때 필수적으로 사용
    
> 주의: 위의 코드는 DBMS(Database Management system)에 따라 실행했을 때 에러가 날 수 있기 때문에     
> 날짜나 문자 타입의 경우에는 산술 연산자를 사용하지 않는 것이 좋습니다.

```python
# dteday 값에 2를 더하고, weekday에 3을 더해보기

mysql("""  

SELECT dteday, dteday + 2, weekday , weekday + 3
FROM bike

""")
```

```python
# 온도(temp)에 +2, -2, *2, /2 추가해보기

mysql("""
 
SELECT temp
    , temp + 2 
    , temp - 2 
    , temp * 2 
    , temp / 2 
FROM bike

""")
```
