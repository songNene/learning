# 특정 데이터 조건과 별칭

## 특정 데이터 조건주기(LIKE, Wildcard)
### LIKE
- 조회 조건 값이 명확하지 않을 때 특정 내용을 찾을 수 있다.
- 조건에 문자나 숫자를 포함할 수 있다.
- 주로 Wildcard와 함께 사용한다.

### Wildcard
- `%` : 조건을 포함하는 모든 문자를 의미합니다. (0개 이상의 문자)
- `_` : 한 글자를 의미합니다.
    - `_`의 개수에 따라 뒤에 오는 문자열의 개수를 정할 수 있습니다.
    - 예: `to_`라고 지정하면 `tom`은 추출되지만 `tomi`는 추출되지 않습니다.

````python
# hr이 0이고, weekday에 'ur'이란 단어가 들어간 데이터를 추출

mysql("""

SELECT *
FROM bike
WHERE hr = 0
AND weekday LIKE '%ur%'

""")
````

````python
# hr이 0이고, weathersit이 'Cl'로 시작하는 데이터를 추출

mysql("""

SELECT *
FROM bike
WHERE hr = 0
      AND weathersit LIKE 'Cl%'

""")
````

````python
# hr이 0이고, mnth이 'ma?'인 데이터를 확인

mysql("""

SELECT *
FROM bike
WHERE hr = 0
      AND mnth LIKE 'ma_'

""")

````

````python
# hr이 0이고, mnth이 'ma?'이며 weekday에 'ur'이란 단어가 들어간 데이터를 추출

mysql(""" 

SELECT *
FROM bike
WHERE hr = 0
      AND mnth LIKE 'ma_'
      AND weekday LIKE '%ur%'

""")
````

### 별칭 사용하기(Alias)
- 임시로 컬럼명 변경시 별칭 사용
- 별칭은 컬럼명 바로 뒤 작성
- 컬럼명과 별칭 사이에는 `AS`를 작성 `AS`는 생략가능
- 특수문자, 대소문자, 공백 사용하려면 `""` 또는 `''`로 묶어줘야한다.
- `SELECT`, `FROM` 절에서 사용
- 뒷부분에서 나올 조인과 서브쿼리에서 테이블과 서브쿼리를 호출하기 위해 **필수로 사용**

````python
# emp와 hum를 추출하고 hum에 'humidity'란 별칭주기

mysql("""

SELECT temp, hum AS 'humidity'
FROM bike
LIMIT 5

""")
````

````python
# bike테이블 이름을 'b'로 지정, b 테이블의 temp와 hum를 추출

mysql(""" 

SELECT b.temp, b.hum AS 'humidity'
FROM bike AS b
LIMIT 5

""")
````

#  데이터 정렬과 중복 제거
## 데이터 정렬하기(ORDER BY)
- SQL 구문 하단에 작성
- 데이터 추출시, 테이블에 입력된 순서대로 출력되는 것이 기본
- 데이터가 오름차순, 내림차순으로 정렬된 형태가 필요할 때가 있다.
- 오름차순 : 컬럼명 뒤 ASC 적기, ASC를 생략해도 기본으로 오름차순으로 정렬됨
- 내림차순 : 컬럼명 뒤 DESC 적기
- SELECT 구분에 지정한 컬럼의 순서를 숫자로 입력해도 실행 가능

````python
# dteday, temp, cnt 컬럼을 추출하고, cnt가 높은 순서대로 100개만 정렬 
# 순서가 같은경우 temp가 낮은 순서대로 정렬됨

mysql("""

SELECT dteday, temp, cnt
FROM bike
ORDER BY cnt DESC, temp
LIMIT 100

""")
````

````python
# 이번에는 컬럼 순서를 입력하여 정렬

mysql("""

SELECT dteday, temp, cnt
FROM bike
ORDER BY 3 DESC, 2
LIMIT 100

""")
````

## 중복제거하기(DISTINCT)
- 출력에서 고유한 값을 반환기 위한 키워드
- `DISTINCT`는 옆에 온 모든 컬럼을 고려해 중복을 제거함
- 뒤에서 매울 집계함수에도 사용할 수 있다.

````python
# weekday, weathersit 컬럼에서 중복을 제거

mysql(""" 

SELECT DISTINCT weekday, weathersit
FROM bike

""")
````

````python
# yr, weekday, weathersit 컬럼에서 중복을 제거

mysql(""" 

SELECT DISTINCT yr, weekday, weathersit
FROM bike


""")
````

# 조건문 활용하기(IF, CASE WHEN ~ THEN ~)
## IF(조건, 참일 때 반환 값, 거짓일 때 반환 값)
- 단일 조건문
- 엑셀의 `IF`함수와 동일한 역할
- SELECT, WHERE 절에서 사용 가능
- IF 내에 조건을 줄 떄 함수와 함께 사용 가능
- 참고 : pandasql에서는 if함수를 iif로 표현

````python
# 1990년 이전과 이후에 태어난 고객 분리하기
# 이해하기 쉽게 '고객명', '생년월일', '1990년 기준'이라는 별칭 사용

mysql("""
 
SELECT customerName AS "고객명", 
      birthday AS "생년월일", 
      iif(birthday >= '1990-01-01' , '이후', '이전') AS "1990년 기준"
FROM customers

""")
````

## CASE WHEN ~ THEN
- 다중 조건문
- 조건에 따른 값을 다르게 출력하고 싶을 때 사용
- 조건에 따라 기존에 있는 컬럼을 이용해 새로운 컬럼을 만들기도 함
- 작성순서
``` sql
CASE WHEN '조건1' THEN '조건1 반환값' 
     WHEN '조건2' THEN '조건2 반환값' 
     ELSE '만족하는 조건이 없을 때 반환 값'
     END
```

- CASE문은 WHEN의 조건을 순차적으로 체크하고 조건이 충족되면 THEN에 명시된 값을 반환
    - WHEN과 THEN은 한 쌍으로 존재합니다.
    - WHEN과 THEN은 여러 개가 존재할 수 있습니다.
    - ELSE는 앞의 조건들에 모두 해당하지 않았을 때 정해줄 기본 값입니다.
    - ELSE가 없고 조건에 맞는 값이 없으면 NULL값을 반환합니다.
- CASE WHEN, THEN, END를 먼저 작성한 후에 조건을 적으면 문법 실수를 줄일 수 있다.

````python
# 고객 한도를 5만원, 십만원 단위로 분리하기
mysql("""
SELECT customerName, creditLimit,
      CASE WHEN creditLimit <= 50000 THEN '5만원이하'
      WHEN creditLimit <= 100000 THEN '10만원이하'
      ELSE '10만원초과'
      END AS '고객한도분리'
FROM customers

""")
````

# 데이터 타입 변환하기(CAST)
## CAST
- 데이터의 타입을 변환하는 함수
- 작성순서
```sql
CAST (컬럼 AS 변환할 타입)
```
- 변환할 타입에는 int, float, varchar, datetime 등이 있다.

````python
# 데이터 bike의 정보를 확인
# 현재 yr은 int 타입

bike.info()
````

````python
# 먼저 2011년 데이터 5개만 추출

mysql("""
SELECT *
FROM bike
WHERE yr = 2011 
LIMIT 5
""")
````

````python
# 현재 int 타입인 yr을 varchar 타입으로 변경

mysql("""
SELECT CAST(yr AS varchar) AS yr
FROM bike
LIMIT 5
""")
````

````python
# yr을 문자열타입으로 변환한 후 2011년 데이터 5개만 추출
# [참고] 보통 해당 타입을 지정하지 않으면 에러가 나지만 데이터베이스 시스템에 따라 타입에 상관 없이 데이터를 추출할 수 있다.

mysql("""
SELECT CAST(yr AS varchar) AS yr
FROM bike
WHERE CAST(yr AS varchar) = '2011'
LIMIT 5
""")
````
