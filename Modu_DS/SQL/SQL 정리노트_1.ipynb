# 데이터와 데이터베이스의 개념

## 데이터 
데이터는 모든 기록들이다.
* 과거에 비해 데이터 양과 복잡성 증대
* 대이터를 저문적으로 봐야하는 직군 증가

## 데이터베이스
데이터가 모여져 있는 공간으로 구조적인 방식으로 관리되는 데이터 집합
* 관리해야하는 데이터들을 목적에 맞게 저장
* 권한에 따라 수정, 검색 등 다룰 수 있는 범위가 다름

## 데이터 베이스 특징
DB는 중복된 데이터를 없애고, 자료를 구조화하여, 효율적인 처리를 할 수 있도록 관리해야한다.
1. 실시간 접근성 : 사용자의 질의에 대하여 즉시 처리하고 응답
2. 지속적인 변화 : 삽입, 삭제, 갱신으로 항상 최신의 데이터 유지하고 DB가 저장하고 있는 내용은 그 당시의 상태
3. 동시공유 : 다수의 사용자가 동시에 같은 내용의 데이터 사용 가능
4. 내용에 대한 참조 : DB에 있는 데이터를 참조할 때 사용자가 요구하는 데이터 내용으로 데이터를 찾을 수 있다. (주소나 행 위치가 아님)

## 관계형 데이터 베이스
* RDB(Relational Database)
* 2차원 테이블 (행,열)로 구성되어 데이터를 정의 및 설명하는 모델
* 구조화된 데이터는 SQL을 통해 조회가능
* 일관성, 안적정, 신뢰성이 있음

# 테이블

## 테이블 개념
- 관계형데이터에서 자료의 구조를 2차원의 표로 나타냄
- 행/열의 형태로 구성 및 관리
- key를 지정해 원하는 데이터를 빠르고 쉽게 찾을 수 있다.

## 행과 열의 개념
1. 행 
    - 가로를 뜻한다.
    - Row 또는 Record라고 불린다.
2. 열
    - 세로를 뜻한다.
    - column 또는 Attribute (속성) 이라고 불린다.

# 스키마와 데이터타입

## 스키마

- 테이블의 뼈대를 구성
- 테이블에 대한 정보가 있다.
- 테이블을 정의할 때 칼럼별로 저장할 수 있는 타입까지 명시해야한다.

## 데이터 타입

- 데이터베이스 테이블에는 특정 데이터 유형이 있는 여러 열이 포함되어있다.
- 프로그래밍 관점에서 데이터를 식별하기 위한 분류
- 문자열, 숫자, 날짜 등의 속성이 있다.
- 데이터 타입에 따라 사용하는 SQL 구문이 다르다.
    - 문자열 : 사칙연산 할 수 없음
    - 날짜 : 날짜 함수는 따로 존재하며 사용시 `""` 따옴표를 사용해야한다.
    - 숫자 : lower() 함수는 사용이 불가하다.
